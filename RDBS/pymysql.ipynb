```python

cursor = connection.cursor()

cursor.execute("CREATE DATABASE mysales_data")

cursor.execute("USE mysales_data")
        
mql = """CREATE TABLE customers (
   FIRST_NAME  CHAR(20) NOT NULL,
   address VARCHAR(255))"""

try:
    cursor.execute(mql)
    
    print("Task added successfully")
except:
    print("Oops! Something wrong")
    
sql = "INSERT INTO customers (FIRST_NAME, address) VALUES (%s, %s)"
val = [
  ('Peter', 'Lowstreet 4'),
  ('Amy', 'Apple st 652'),
  ('Hannah', 'Mountain 21'),
  ('Michael', 'Valley 345'),
  ('Sandy', 'Ocean blvd 2'),
  ('Betty', 'Green Grass 1'),
  ('Richard', 'Sky st 331'),
  ('Susan', 'One way 98'),
  ('Vicky', 'Yellow Garden 2'),
  ('Ben', 'Park Lane 38'),
  ('William', 'Central st 954'),
  ('Chuck', 'Main Road 989'),
  ('Viola', 'Sideway 1633')
]
try:
    cursor.executemany(sql,val)
    print("Task added successfully")
    connection.commit()
except:
    print("Oops! Something wrong")



connection.close()
```
